## Part D â€” Detailed EDA & Visualization Audit

### 1. Data Health & Descriptive Profiling

**What we examine**

- Overall row counts, number of species, key numerical and categorical variables  
- Missing value patterns and basic sanity checks (weights, heights, etc.)  

**Findings**

- âœ… Core descriptive statistics (mean, median, std, min, max) are correctly computed.  
- âœ… Missing values are visible at column level, enabling quick assessment.  
- âš ï¸ There is limited use of **data health KPIs** such as:  
  - proportion of invalid / extreme values per key variable,  
  - consistency checks (e.g. weight vs length ranges per species).  

**Recommendation**

- Introduce a concise **Data Health Summary** table with:  
  - row count, distinct species count, missingness per variable,  
  - simple business rules (e.g. weight > 0, reasonable upper limits).  

---

### 2. Distribution Analysis

**What we examine**

- Histograms, kernel density estimates (KDE), boxplots, and empirical distributions of key variables.  

**Findings**

- âœ… Distribution plots are present and correctly configured.  
- âœ… Skewness is visually detectable in some variables.  
- âš ï¸ Skewness, kurtosis or normality are not **quantified**, only visually inferred.  
- âš ï¸ The implications for modelling (e.g. need for log transform, binning) are not explicitly stated.  

**Recommendation**

- Compute and display **skewness and kurtosis** alongside distribution plots.  
- Add explicit interpretations:  
  > â€œWeight_kg is right-skewed; consider log-transforming for linear models.â€  

---

### 3. Relationships & Correlations

**What we examine**

- Correlation matrices, scatter plots, pairplots, and other relationship diagnostics.  

**Findings**

- âœ… Correlation heatmaps highlight linear relationships between key numerical variables.  
- âœ… Some pairwise relationships are explored.  
- âš ï¸ The most important relationships (from a modelling perspective) are not **ranked** or prioritised.  
- âš ï¸ Non-linear relationships are not explicitly tested (e.g. polynomial patterns, thresholds).  

**Recommendation**

- Produce a **â€œTop 5 strongest correlationsâ€** summary table and add commentary.  
- Explicitly link observed patterns to **feature engineering decisions** (e.g. interaction terms).  

---

### 4. Segmentation & Subgroup Exploration

**What we examine**

- Whether EDA distinguishes between key segments such as species, region, environment, or weight bands.  

**Findings**

- âš ï¸ Most analysis is performed at **global level** (all animals aggregated).  
- â— This under-utilises the richness of the dataset and may obscure segment-specific insights.  

**Recommendation**

- For each key question, provide both:  
  - an **overall view** and  
  - a **segmented view** (e.g. per species).  

Example:
- Distribution of weight overall, then per species.  
- Length vs weight relationship overall, then coloured by species.  

---

### 5. Outlier & Anomaly Detection

**What we examine**

- How atypical values are surfaced and handled.  

**Findings**

- âœ… Outliers are visible on boxplots and scatterplots.  
- âš ï¸ There is no **systematic thresholding** or labelling for outliers.  
- âš ï¸ No clear framing of which anomalies are **data errors** vs valid but rare cases.  

**Recommendation**

- Introduce a simple **rule-based outlier flagging** (e.g. beyond 3Ã—IQR).  
- Summarise:  
  - how many records are flagged as potential outliers,  
  - decision recommendation: inspect, cap, remove, or keep as-is.  

---

### 6. Visualization Design & Storytelling

**What we examine**

- Titles, subtitles, axis labels, annotations, colour usage, and ordering of plots.  

**Findings**

- âœ… Visuals are readable for technical audiences.  
- âš ï¸ Titles are sometimes descriptive but not **insight-focused**.  
- âš ï¸ Few charts explicitly highlight a **takeaway sentence**.  
- âš ï¸ No consistent **visual system** (colours, fonts, spacing) that would feel â€œproductisedâ€ or â€œbrand-alignedâ€.  

**Recommendation**

- Adopt a simple **visual style guide** (colour palette, font sizes, annotation style).  
- For key plots, add a one-line **â€œKey Insightâ€** directly below the figure.  
- Reorder plots to support a clear storyline, e.g.:  
  1. Data coverage and health,  
  2. Core distributions,  
  3. Key relationships,  
  4. Segment differences,  
  5. Outliers & edge cases.

## Code Appendix & Reproducible Templates

> â„¹ï¸ The following examples illustrate how to structure EDA code for **clarity, reuse, and governance**.

````python
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path

# Example paths (adapt to your config_utils or paths_utils)
DATA_CLEAN = Path("data/cleaned/clean_animal_data.csv")


def load_clean_dataset(path: Path = DATA_CLEAN) -> pd.DataFrame:
    """Load the main cleaned dataset for EDA."""
    df = pd.read_csv(path)
    print(f"âœ… Data loaded â€” shape={df.shape}, columns={list(df.columns)}")
    return df


def plot_distribution(df: pd.DataFrame, column: str):
    """Simple reusable distribution plot (hist + KDE placeholder)."""
    fig, ax = plt.subplots(figsize=(6, 4))
    df[column].hist(ax=ax, bins=30)
    ax.set_title(f"Distribution of {column}")
    ax.set_xlabel(column)
    ax.set_ylabel("Count")
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    df_clean = load_clean_dataset()
    # Example call
    if "Weight_kg" in df_clean.columns:
        plot_distribution(df_clean, "Weight_kg")
````
